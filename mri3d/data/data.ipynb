```python
from empatches import EMPatches
import numpy as np
import torch
emp = EMPatches()

a = torch.rand(1,3,192,144,144)
def extract_patches_5d(source, patchsize=96, overlap=0.25, stride=None, emp=emp,vox=True):
    assert len(source.shape) == 5
    patches, indices = emp.extract_patches(source[0,0,:,:,:],patchsize=patchsize, overlap=overlap, stride=stride, vox=vox)
    a=torch.stack(patches).unsqueeze(1).unsqueeze(1).repeat(1,1,3,1,1,1)
    return a

patches = extract_patches_5d(a, patchsize=96, overlap=0.25, stride=None, vox=True,emp=emp)
```
